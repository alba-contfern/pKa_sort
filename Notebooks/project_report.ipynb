### Limitations:
<p>This program is limited to the information found on PubChem. For example, since in the page for 'Water' doesn't explicitely contain a pKa page embedded in the HTML, it will give an error. Furthermore, the main function requires a lot of CPU. We tried to shorten the function that looks up the pka value on pubchem, but the functionnalities we could take off didn't improve the speed of the function, so we left the original one. As a result, when more than three molecules are input to compare, it takes a long time to run and get the result. The user is warned about this issue:


    
Consequently, we removed the functionnality of imaging so that the function can still be used to sort the pKas:
    

    
The program also has issues with molecules that have multiple acidic groups, for example aminoacids, and it only returns either the amine or the carboxilic acid pKa. 
With more time, it would've also been interesting to modify the code in order to not only get the generated image, but also highlight all of the acidic groups of the molecule, instead of only the hydroxyls. </p>



### Challenges faced

<p>We lost a lot of time (several days) trying to program ourselves the function that finds the pKas of the compounds by parsing the HTML code of the PubChem pages. However, even with the help of the assistants we couldn't manage, as the information was too deeply embedded and not accessible through parsing with 'beautiful soup'. The IPA didn't contain the pKa information, and the code was unsuccessful (see Notebooks/Failed_verion.ipynb). Instead, we used a function that we found from another programmer that used Python XML as to parse the XML format of the PubChem page, and that worked well. We were also unable to generate all of the molecule images at once, and the object wasn't subscriptable to a list so we were having trouble with that. Finally we found found the solution with an rdkit function (MolsToGridImage). While we coded the interface we also faced trouble to generate the images on it, and that's why it's only available for three molecules at most. The images first had to be converted to png to be compatible with the interface output. Then, the number limit is due to the fact that for a higher number, the size of the images made it so they discentered from the display.   </p>
